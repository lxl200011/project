```python
def formatcol(df):
    df.columns=df.columns.str.replace(u"\\(.*?\\)",'').str.strip().str.lower().str.replace(' ', '_').str.replace('/', '').str.replace('?','')
    return(df)
formatcol(df)
def addopp(df):
    newname1=[(i,'opp_'+i) for i in df.iloc[:,8:19].columns.values]
    df.rename(columns=dict(newname1),inplace=True)
    newname2=[(i,'opp_'+i) for i in df.iloc[:,22:42].columns.values]
    df.rename(columns=dict(newname2),inplace=True)
    newname3=[(i,'opp_'+i) for i in df.iloc[:,44:46].columns.values]
    df.rename(columns=dict(newname3),inplace=True)
    return df
addopp(df)    
df.columns
df.dtypes
```

```python
colname={'client_type':{'New':0,'Existing':1},'opp_is_business_seasonal':{'Open full time': 100000000,'Open seasonally': 100000001,'Open part time': 100000002, 'Not started': 100000003},'opp_bank_account_type':{'Both':100000002,'Personal':100000000,'Business':100000001},'opp_is_homeowner':{'No':100000000,'Yes':100000001},'opp_acct_record':{'None':1,'2 - Some (Informal)':2,'3 - Regular (Formal)':3,'4 - Professional':4},'opp_company_type':{'Limited Liability Company (LLC)':1,'Corporation':2,'Sole Proprietorship/Individual':3},'business_location_type':{'Office Building':4,'Home':1,'Market':3,'Street':5,'Storefront':2,'Other':6},'opp_has_state_tax_lien':{'Yes':1,'No':0},'opp_has_federal_tax_lien':{'Yes':1,'No':0},'opp_has_civil_judgement':{'Yes':1,'No':0},'coborrower_status':{'Yes':1,'No':0}}
def changetype(df,colname):
    for x,y in colname.items():
        df[x]=df[x].astype('category').map(y)
    return df        
changetype(df,colname)               
```

```python
variablename1='model_target_total_balance'
loanamount='total_loan_amount'
balance='total_balance'
def maketargvar(df,variablename1,loanamount,balance):
    df[variablename]=np.where((df[balance]/df[loanamount])<=0.05,1,0)
    df[variablename]=df[variablename].astype('category')
maketargvar(df,variablename1,loanamount,balance)     
df.info()
```

```python
variablename2='model_date_opp_established_date'
clodate='closing_date'
estdate='opp_date_bz_established'
def makedatevar(df,variablename2,clodate,estdate):
    df[clodate]=pd.to_datetime(df[clodate])
    df[estdate]=pd.to_datetime(df[estdate])
    df[variablename2]=df[clodate]-df[estdate]
    df[variablename2]=df[variablename2].dt.days
makedatevar(df,variablename2,clodate,estdate)     
df.info()
```

```python
file_output3="internship3.json"
def makejson(df,file_output3):
    result=df.to_json(orient='records')
    parsed=json.loads(result)
    with open(file_output3,'w',encoding='cp1252')as jsonoutfile:
        jsonString=json.dumps(parsed,indent=4)
        jsonoutfile.write(jsonString)
makejson(df,file_output3)  
```
